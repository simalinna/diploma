# Обучение

```python
import pandas as pd
from matplotlib import pyplot as plt
```

```python
stats_df = pd.read_json('../experiments/exp1/stats.json')
```

### Значение функции ошибки на каждой эпохе

```python
plt.plot(stats_df['epoch'], stats_df['loss'], color='m', label='Обучение')
plt.plot(stats_df['epoch'], stats_df['val_loss'], color='darkorange', label='Валидация')

plt.legend(fontsize=10)

plt.xlabel('Эпоха')
plt.ylabel('Функция ошибки')

plt.grid(True)

plt.show()
```

### Значение компонент функции ошибки на каждой эпохе

```python
plt.plot(stats_df['epoch'], stats_df['vicreg_loss'], color='m', label='Обучение')
plt.plot(stats_df['epoch'], stats_df['val_vicreg_loss'], color='b', label='Валидация')

plt.legend(fontsize=10)

plt.xlabel('Эпоха')
plt.ylabel('Компоненты функции ошибки')

plt.grid(True)

plt.show()
```

```python
plt.plot(stats_df['epoch'], stats_df['ce_loss'], color='m', label='Обучение')
plt.plot(stats_df['epoch'], stats_df['val_ce_loss'], color='b', label='Валидация')

plt.legend(fontsize=10)

plt.xlabel('Эпоха')
plt.ylabel('Компоненты функции ошибки')

plt.grid(True)

plt.show()
```
